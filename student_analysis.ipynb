```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#df = pd.read_csv("student-mat.csv", sep=';'): Reads the CSV file named 'student-mat.csv' into a pandas DataFrame called df. The sep=';' argument specifies that the data in the file is separated by semicolons, not commas.
df = pd.read_csv("student-mat.csv", sep=';')

print("Dataset Shape:", df.shape) #Prints the number of rows and columns in the DataFrame.
print(df.dtypes) #Prints the data type of each column in the DataFrame.

print("Missing values:\n", df.isnull().sum()) #Checks for and prints the count of missing values in each column.
df = df.drop_duplicates() #Removes any duplicate rows from the DataFrame.

avg_g3 = df["G3"].mean() #Calculates the average of the 'G3' (final grade) column.
print(f"Average Final Grade:{avg_g3:.2f}") #Prints the calculated average final grade, formatted to two decimal places.

top_students = df[df['G3']>15].shape[0] #Counts the number of students who scored above 15 in their final grade (G3).
print(f"Students scoring above 15: {top_students}") #Prints the count of top students.

gender_performance = df.groupby('sex')['G3'].mean() # Calculates the average final grade ('G3') grouped by 'sex' (gender).
print("\nAverage score by gender:\n", gender_performance) #Prints the average final grade for each gender.

correlation = df['studytime'].corr(df['G3']) #Calculates the Pearson correlation coefficient between 'studytime' and 'G3'.
print(f"\nCorrelation between Study Time and G3: {correlation:.2f}") #Prints the correlation, formatted to two decimal places.

sns.set_theme(style="whitegrid") # Sets the seaborn plotting style to 'whitegrid', which is useful for plots with grids.

plt.figure(figsize=(8, 5)) # Creates a new figure for the plot with a specified size.
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black') # Creates a histogram of the 'G3' column with 10 bins, 'skyblue' color, and black edges.
plt.title("Distribution of Final Grades (G3)") #Sets the title of the histogram.
plt.xlabel("Grade") #Labels the x-axis
plt.ylabel("Frequency")  #Labels the y-axis.
plt.show() # Displays the histogram.

plt.figure(figsize=(8, 5)) #Creates another new figure.
sns.scatterplot(x='studytime', y='G3', data=df, color='green') #Generates a scatter plot using seaborn, showing 'studytime' on the x-axis and 'G3' on the y-axis, with points in green.
plt.title("Study Time vs Final Grade") #Sets the title of the scatter plot.
plt.show() #Displays the scatter plot.

plt.figure(figsize=(6, 5)) #Creates a third new figure.
gender_performance.plot(kind='bar', color=['pink', 'lightblue']) #Creates a bar plot from the gender_performance Series, using 'pink' and 'lightblue' for the bars.
plt.title("Average Score: Male vs Female") #Sets the title of the bar plot.
plt.ylabel("Average G3") #Labels the y-axis.
plt.xticks(rotation=0) #Sets the rotation of x-axis labels to 0 degrees.
plt.show() #Displays the bar plot.
```
